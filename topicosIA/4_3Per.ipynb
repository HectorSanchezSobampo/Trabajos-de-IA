## ¿Qué es un Perceptrón?

El Perceptrón es uno de los algoritmos de aprendizaje automático más antiguos y sencillos, desarrollado por **Frank Rosenblatt en 1957** en la Universidad de Cornell. Fue el primer modelo de red neuronal artificial en ser implementado. Su concepción fue un hito importante, ya que sentó las bases para el desarrollo posterior de las redes neuronales y el campo del aprendizaje profundo, aunque en sus inicios solo podía resolver problemas linealmente separables.

### Componentes Clave del Perceptrón

Un Perceptrón simula el funcionamiento básico de una neurona biológica y consta de varios elementos fundamentales:

*   **Entradas (Inputs):** Son los valores numéricos que el Perceptrón recibe. Cada entrada representa una característica o "feature" de los datos. Por ejemplo, en un problema de clasificación de correos electrónicos, las entradas podrían ser la frecuencia de ciertas palabras.

*   **Pesos (Weights):** Cada entrada tiene un peso asociado. Los pesos son valores numéricos que indican la importancia de cada entrada para la decisión final del Perceptrón. Durante el proceso de entrenamiento, estos pesos se ajustan para que el modelo aprenda a clasificar correctamente.

*   **Sesgo (Bias) o Umbral (Threshold):** El sesgo es un valor adicional que se suma al producto de las entradas por sus pesos. Funciona como un intercepto en una regresión lineal y permite que la función de activación se desplace, lo que ayuda al Perceptrón a aprender patrones que no pasan por el origen. Tradicionalmente, en la formulación original, se hablaba de un umbral que las sumas ponderadas debían superar. Ambos conceptos cumplen una función similar en la toma de decisión.

*   **Suma Ponderada (Weighted Sum):** Es el corazón del Perceptrón. Se calcula multiplicando cada entrada por su peso correspondiente y sumando todos estos productos, a lo que finalmente se le añade el sesgo. Matemáticamente, esto se expresa como: `Z = (x1 * w1) + (x2 * w2) + ... + (xn * wn) + bias`, donde `x` son las entradas y `w` son los pesos.

*   **Función de Activación (Activation Function):** Una vez calculada la suma ponderada, esta pasa por una función de activación. Para el Perceptrón original, la función de activación más común es la **función escalón (step function)** o función de activación binaria. Esta función produce una salida binaria (0 o 1, o -1 o 1) basada en si la suma ponderada supera o no un cierto umbral (o si es mayor o menor que cero, si el sesgo ya está incorporado).

### ¿Cómo Funciona el Perceptrón para la Clasificación Binaria?

El Perceptrón está diseñado para resolver problemas de **clasificación binaria**, es decir, para categorizar datos en una de dos clases posibles (por ejemplo, "sí" o "no", "verdadero" o "falso", "clase A" o "clase B"). El proceso de toma de decisiones funciona de la siguiente manera:

1.  **Recibir Entradas:** El Perceptrón recibe un conjunto de entradas (características) para un dato específico.
2.  **Calcular Suma Ponderada:** Multiplica cada entrada por su peso asignado y suma estos productos, añadiendo el sesgo. Esto da como resultado un valor único (`Z`).
3.  **Aplicar Función de Activación:** El valor `Z` se introduce en la función de activación. Si la función de activación escalón se usa, la salida será:
    *   `1` (o "Clase Positiva") si `Z` es mayor o igual al umbral (o mayor o igual a 0 si el sesgo está incluido).
    *   `0` (o "Clase Negativa") si `Z` es menor que el umbral (o menor que 0 si el sesgo está incluido).

**Ejemplo:** Imagina que queremos clasificar si una fruta es una manzana (1) o no (0) basándonos en su peso y color. El Perceptrón recibiría el peso (`x1`) y un valor numérico para el color (`x2`). Multiplicaría `x1` por su peso `w1`, `x2` por `w2`, sumaría estos productos y el sesgo. Si el resultado de la suma ponderada es suficientemente alto (indicando que es más probable que sea una manzana), la función de activación emitirá un `1`; de lo contrario, emitirá un `0`.

El Perceptrón aprende ajustando sus pesos y el sesgo durante el entrenamiento, basándose en los errores de clasificación que comete. Si clasifica un dato incorrectamente, los pesos se ajustan ligeramente para que la próxima vez la suma ponderada y la salida se acerquen a la clasificación correcta. Este proceso iterativo continúa hasta que el Perceptrón puede clasificar la mayoría de los datos correctamente, o hasta que se alcanza un número máximo de iteraciones.

Es importante destacar que el Perceptrón original solo puede clasificar correctamente datos que son **linealmente separables**, es decir, datos que pueden ser divididos en dos clases por una línea recta (en 2D), un plano (en 3D) o un hiperplano (en dimensiones superiores).

## Implementación del Perceptrón

### Subtask:
Crear una clase de Perceptrón en Python que incluya métodos para inicializar los pesos y el sesgo, calcular la suma ponderada, aplicar la función de activación y realizar la predicción.


**Reasoning**:
The subtask requires implementing a Perceptron class in Python. This code block will define the class with the specified methods for initialization, weighted sum calculation, activation function, and prediction, along with detailed comments as requested in the instructions.



````python
import numpy as np

class Perceptron:
    """
    Implementa un Perceptrón simple para clasificación binaria.

    Atributos:
        learning_rate (float): La tasa de aprendizaje para actualizar los pesos.
        n_epochs (int): El número de épocas para el entrenamiento.
        weights (np.array): Los pesos del perceptrón, inicializados con ceros.
        bias (float): El sesgo del perceptrón, inicializado en cero.
    """

    def __init__(self, learning_rate: float = 0.01, n_epochs: int = 100):
        """
        Inicializa el Perceptrón con una tasa de aprendizaje y número de épocas.
        Los pesos se inicializan como un array de ceros y el sesgo como cero.

        Parámetros:
            learning_rate (float): Tasa de aprendizaje para el ajuste de pesos. Por defecto es 0.01.
            n_epochs (int): Número de iteraciones de entrenamiento. Por defecto es 100.
        """
        self.learning_rate = learning_rate
        self.n_epochs = n_epochs
        self.weights = None  # Se inicializarán con ceros al entrenar, según el número de características
        self.bias = 0.0
        print("Perceptron class initialized with learning_rate: {self.learning_rate} and n_epochs: {self.n_epochs}")

    def _weighted_sum(self, X: np.array) -> float:
        """
        Calcula la suma ponderada de las características de entrada y los pesos,
        y luego añade el sesgo.

        Parámetros:
            X (np.array): Vector de características de entrada.

        Retorna:
            float: La suma ponderada más el sesgo.
        """
        if self.weights is None:
            raise ValueError("Weights have not been initialized. Please train the perceptron first.")
        return np.dot(X, self.weights) + self.bias

    def _step_function(self, weighted_sum: float) -> int:
        """
        Aplica la función de activación (función escalón).
        Retorna 1 si la suma ponderada es mayor o igual a 0, y 0 en caso contrario.

        Parámetros:
            weighted_sum (float): El resultado de la suma ponderada.

        Retorna:
            int: La salida binaria (0 o 1).
        """
        return 1 if weighted_sum >= 0 else 0

    def predict(self, X: np.array) -> int:
        """
        Realiza una predicción para un vector de características de entrada.
        Calcula la suma ponderada y aplica la función de activación.

        Parámetros:
            X (np.array): Vector de características de entrada.

        Retorna:
            int: La predicción binaria (0 o 1).
        """
        weighted_sum = self._weighted_sum(X)
        prediction = self._step_function(weighted_sum)
        return prediction

# Example of how to use the Perceptron (for demonstration, not part of the class itself)
# perceptron = Perceptron(learning_rate=0.1, n_epochs=10)
# print(perceptron)
````

## Entrenamiento del Perceptrón

### Subtask:
Implementar el algoritmo de entrenamiento del Perceptrón, incluyendo la inicialización de pesos, la iteración a través de los datos de entrenamiento, el cálculo del error y la actualización de pesos y el sesgo. Se almacenarán los valores de los pesos y el sesgo en cada época.


**Reasoning**:
I need to modify the existing `Perceptron` class to include the `fit` method as described in the instructions. This method will implement the training algorithm, including weight initialization, iterative updates, error calculation, and history tracking for weights and bias. After defining the class, I will instantiate and train the Perceptron using the previously generated dataset.



````python
import numpy as np

class Perceptron:
    """
    Implementa un Perceptrón simple para clasificación binaria.

    Atributos:
        learning_rate (float): La tasa de aprendizaje para actualizar los pesos.
        n_epochs (int): El número de épocas para el entrenamiento.
        weights (np.array): Los pesos del perceptrón, inicializados con ceros.
        bias (float): El sesgo del perceptrón, inicializado en cero.
        weights_history (list): Historial de los pesos en cada época de entrenamiento.
        bias_history (list): Historial del sesgo en cada época de entrenamiento.
    """

    def __init__(self, learning_rate: float = 0.01, n_epochs: int = 100):
        """
        Inicializa el Perceptrón con una tasa de aprendizaje y número de épocas.

        Parámetros:
            learning_rate (float): Tasa de aprendizaje para el ajuste de pesos. Por defecto es 0.01.
            n_epochs (int): Número de iteraciones de entrenamiento. Por defecto es 100.
        """
        self.learning_rate = learning_rate
        self.n_epochs = n_epochs
        self.weights = None  # Se inicializarán con ceros al entrenar, según el número de características
        self.bias = 0.0
        self.weights_history = []
        self.bias_history = []
        print(f"Perceptron class initialized with learning_rate: {self.learning_rate} and n_epochs: {self.n_epochs}")

    def _weighted_sum(self, X: np.array) -> float:
        """
        Calcula la suma ponderada de las características de entrada y los pesos,
        y luego añade el sesgo.

        Parámetros:
            X (np.array): Vector de características de entrada.

        Retorna:
            float: La suma ponderada más el sesgo.
        """
        if self.weights is None:
            # This check is primarily for predict method before fitting, but during fit, weights will be initialized.
            raise ValueError("Weights have not been initialized. Please train the perceptron first.")
        return np.dot(X, self.weights) + self.bias

    def _step_function(self, weighted_sum: float) -> int:
        """
        Aplica la función de activación (función escalón).
        Retorna 1 si la suma ponderada es mayor o igual a 0, y 0 en caso contrario.

        Parámetros:
            weighted_sum (float): El resultado de la suma ponderada.

        Retorna:
            int: La salida binaria (0 o 1).
        """
        return 1 if weighted_sum >= 0 else 0

    def predict(self, X: np.array) -> int:
        """
        Realiza una predicción para un vector de características de entrada.
        Calcula la suma ponderada y aplica la función de activación.

        Parámetros:
            X (np.array): Vector de características de entrada.

        Retorna:
            int: La predicción binaria (0 o 1).
        """
        # Ensure X is treated as a 1D array for single predictions if it comes as 2D with one sample
        if X.ndim == 2 and X.shape[0] == 1:
            X = X.flatten()

        weighted_sum = self._weighted_sum(X)
        prediction = self._step_function(weighted_sum)
        return prediction

    def fit(self, X: np.array, y: np.array):
        """
        Entrena el Perceptrón con los datos de entrenamiento.

        Parámetros:
            X (np.array): Matriz de características de entrada (n_muestras, n_características).
            y (np.array): Vector de etiquetas verdaderas (n_muestras,).
        """
        # a. Inicializar los pesos y el sesgo
        n_features = X.shape[1]
        self.weights = np.zeros(n_features) # Inicializar pesos con ceros
        self.bias = 0.0 # Inicializar sesgo en cero

        # c. Crear listas para almacenar el historial de pesos y sesgos
        self.weights_history = []
        self.bias_history = []

        print("Starting Perceptron training...")
        # d. Iterar sobre el número de épocas
        for epoch in range(self.n_epochs):
            # e. Dentro de cada época, iterar sobre cada muestra de entrenamiento
            for i in range(X.shape[0]):
                x_i = X[i]
                y_i = y[i]

                # f. Calcular la predicción
                y_pred = self.predict(x_i)

                # g. Calcular el error
                error = y_i - y_pred

                # h. Actualizar pesos y sesgo si hay error
                if error != 0:
                    self.weights += self.learning_rate * error * x_i
                    self.bias += self.learning_rate * error

            # i. Almacenar los pesos y el sesgo actuales al final de la época
            self.weights_history.append(self.weights.copy())
            self.bias_history.append(self.bias)
            # print(f"Epoch {epoch+1}/{self.n_epochs} - Weights: {self.weights}, Bias: {self.bias}")
        print("Perceptron training finished.")

# Instantiate and train the Perceptron
perceptron = Perceptron(learning_rate=0.1, n_epochs=10)
perceptron.fit(X, y)

print(f"\nFinal trained weights: {perceptron.weights}")
print(f"Final trained bias: {perceptron.bias}")
````

First, I need to define the `plot_decision_boundary` helper function that will be used to visualize the Perceptron's decision boundary at different stages of training. This function will calculate the line based on the weights and bias and plot it on a given matplotlib axis.



````python
import numpy as np
import matplotlib.pyplot as plt

# Helper function to plot the decision boundary
def plot_decision_boundary(ax, weights, bias, x_min, x_max, color, linestyle, label):
    """
    Calcula y dibuja la línea de decisión del Perceptrón.
    La línea de decisión está definida por w1*x1 + w2*x2 + bias = 0.
    """
    if weights[1] == 0: # Handle vertical line case
        if weights[0] == 0: # No clear decision boundary if both weights are zero
            line_x = np.linspace(x_min, x_max, 100)
            line_y = np.full_like(line_x, -bias) # A horizontal line if bias is non-zero
        else:
            # If w1 != 0 and w2 == 0, then w1*x1 + bias = 0 => x1 = -bias / w1
            ax.axvline(x=-bias/weights[0], color=color, linestyle=linestyle, label=label)
            return
    else:
        # x2 = (-weights[0]*x1 - bias) / weights[1]
        line_x = np.linspace(x_min, x_max, 100)
        line_y = (-weights[0] * line_x - bias) / weights[1]

    ax.plot(line_x, line_y, color=color, linestyle=linestyle, label=label)

# Now, let's proceed with plotting the decision boundary evolution and parameter updates.
# We will use the 'perceptron' object trained in the previous step.

# Get the min and max values for the features to define plot limits
x1_min, x1_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
x2_min, x2_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

# 1. Visualización de la evolución de la línea de decisión
# Seleccionamos algunas épocas clave para visualizar
epochs_to_plot = [0, 2, 5, perceptron.n_epochs - 1] # Inicial, algunas intermedias y final

fig_db, axes_db = plt.subplots(1, len(epochs_to_plot), figsize=(5 * len(epochs_to_plot), 5), sharex=True, sharey=True)
fig_db.suptitle('Evolución de la Línea de Decisión del Perceptrón', fontsize=16)

for i, epoch_idx in enumerate(epochs_to_plot):
    ax = axes_db[i]

    # Plot original data points
    ax.scatter(X[y == 0, 0], X[y == 0, 1], color='red', marker='o', label='Clase 0')
    ax.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', marker='x', label='Clase 1')

    # Get weights and bias for the current epoch
    current_weights = perceptron.weights_history[epoch_idx]
    current_bias = perceptron.bias_history[epoch_idx]

    # Plot the decision boundary
    plot_decision_boundary(ax, current_weights, current_bias, x1_min, x1_max, 'green', '-', 'Línea de Decisión')

    ax.set_title(f'Época {epoch_idx + 1}')
    ax.set_xlabel('Característica 1')
    ax.set_ylabel('Característica 2')
    ax.set_xlim(x1_min, x1_max)
    ax.set_ylim(x2_min, x2_max)
    ax.grid(True)
    if i == 0:
        ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.9])
plt.show()

# 2. Visualización de la evolución de los pesos
plt.figure(figsize=(10, 5))
weights_history_arr = np.array(perceptron.weights_history)
plt.plot(range(1, perceptron.n_epochs + 1), weights_history_arr[:, 0], label='Peso w1', marker='o', markersize=4)
plt.plot(range(1, perceptron.n_epochs + 1), weights_history_arr[:, 1], label='Peso w2', marker='x', markersize=4)
plt.title('Evolución de los Pesos durante el Entrenamiento')
plt.xlabel('Época')
plt.ylabel('Valor del Peso')
plt.grid(True)
plt.legend()
plt.show()

# 3. Visualización de la evolución del sesgo
plt.figure(figsize=(10, 5))
plt.plot(range(1, perceptron.n_epochs + 1), perceptron.bias_history, label='Sesgo', color='purple', marker='s', markersize=4)
plt.title('Evolución del Sesgo durante el Entrenamiento')
plt.xlabel('Época')
plt.ylabel('Valor del Sesgo')
plt.grid(True)
plt.legend()
plt.show()
````
