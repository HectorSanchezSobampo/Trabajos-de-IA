### Generar una Gráfica Simple con Matplotlib

Este código utiliza la librería `matplotlib` para crear un gráfico de líneas. Primero, importamos `matplotlib.pyplot` como `plt`. Luego, generamos algunos datos de ejemplo y finalmente usamos `plt.plot()` para dibujar la gráfica y `plt.show()` para mostrarla.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar algunos datos de ejemplo
x = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y = np.sin(x) # Función seno de x

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Seno(x)', color='blue', linestyle='--')
plt.title('Gráfica de la Función Seno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.legend()
plt.savefig('grafica_seno.png') # Guarda la gráfica como un archivo PNG
plt.show()
```
